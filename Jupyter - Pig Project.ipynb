# Tasks to follow
- allow user to roll the a dice. Need a random generator to give us a number between 1 - 6
- We then to need to ask the user if they want to continue to roll the dice. Need a yes or no option, if yes, roll again.
- When they stop their turn we need the score to go to their final score
- Have a limit on the final score of 50. If someone reaches 50, they win and game is ended.

```python
import random #importing random, a module in Python's standard library. The random module provides tools for generating random numbers, choosing random items, and more.
def roll(): #creating a random roll function
    min_value = 1 #input min number the dice will have
    max_value = 6 #input max number the dice will have 
    roll = random.randint(min_value, max_value) #.randint is a function inside the random module. It stands for "random integer." Randint comes after random because it is a function inside random module.
    return roll #signifies the function to return the value of the roll object
```

```python
while True: # a while true loop creates an infinite loop that continues to ask the player for # of players, until number given is valid.
    players = input('Enter the number of players (2 - 4): ') #defining players, using input function that will intake a string reponse. Giving an initial message by already entering something in the input brackets
    if players.isdigit(): #Checks if the input is a valid whole number. If the user enters "abc" or "2.5", this evaluates to False. Goes to 'Invalid'
        players = int(players) #if players is a whole number, int() will convert the string input into a integer
        if 2 <= players <=4: #Checks if the number of players is between 2 and 4 (inclusive).
            break #breaks the infinte while true loop. This happens only if the input passes all validation checks.
        else:
            print('Must be between 2 - 4 players') #If the input is a number (due to isdigit() returning True) but is not between 2 and 4, this message is printed.
    else: print('Invalid, try again.') #If players.isdigit() is False (meaning the input was not a number), this error message is printed.     
        
print(players) #After the loop is exited (because valid input was provided), the program prints the valid number of players.      
```

```python
max_score = 50  #setting the max score of 50, once this score is reached the game will end.
player_scores = [0 for i in range(players)] #creates a list and initializes it with zeros, one for each player. It will eventually hold the scores for all players.
#The range() function generates a sequence of numbers starting from 0 up to (but not including) the value of players. For example, if players = 3, range(players) produces 0, 1, 2.
#for i in range(players): A loop is run for each number in the sequence generated by range(players). For each iteration of the loop, the number 0 is added to the list.
print(player_scores)
```

```python
while max(player_scores) < max_score:
    for player_idx in range(players):
        print('\nPlayer number', player_idx + 1, 'turn has just started!\n')
        print('Your total score is:', player_scores[player_idx],'\n')
        current_score = 0
        
        while True:
            should_roll = input('would you like to roll (y)?')
            if should_roll.lower() != 'y':
                break
            else:
                value = roll()
                if value == 1:
                    print('You rolled a 1! Turn done!')
                    current_score = 0 
                    break
                else:
                    current_score += value
                    print('You rolled a: ', value)
                    
                print('Your score is:', current_score)
                
        player_scores[player_idx] += current_score
        print('Your total score is:', player_scores[player_idx])
```

```python
highest_score = max(player_scores)
winning_idx = player_scores.index(max_score)
print('player number', winning_idx + 1, 'is the winner with a score of:', max_score)
```
